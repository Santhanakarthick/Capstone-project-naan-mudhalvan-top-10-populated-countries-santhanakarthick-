#  **Data Visualization:**

### Bar chart 2024

```python
# Filter data for the year 2024
df_2024 = df[df['year'] == 2024]

# Plotting the bar chart for the year 2024 with different colors for each country and text on each bar
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
bars = plt.bar(df_2024.columns[1:], df_2024.values[0][1:], color=['blue', 'red', 'green', 'orange', 'purple', 'brown', 'pink', 'gray', 'cyan', 'magenta'])
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Population Comparison for the Year 2024')
plt.xticks(rotation=45)
plt.tight_layout()

# Adding text on each bar
for bar, population in zip(bars, df_2024.values[0][1:]):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{population:,}', ha='center', va='bottom')

plt.show()
```

### Bar chart 2050:

```python
# Filter data for the year 2050
df_2050 = df[df['year'] == 2050]

# Plotting the bar chart for the year 2050 with different colors for each country and text on each bar
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
bars = plt.bar(df_2050.columns[1:], df_2050.values[0][1:], color=['blue', 'red', 'green', 'orange', 'purple', 'brown', 'pink', 'gray', 'cyan', 'magenta'])
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Population Comparison for the Year 2050')
plt.xticks(rotation=45)
plt.tight_layout()

# Adding text on each bar
for bar, population in zip(bars, df_2050.values[0][1:]):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{population:,}', ha='center', va='bottom')

plt.show()
```

### Pie Chart 2024:

```python
# here we use pie chart to see population share of 10 countries

# Filter data for the year 2024
df_2024 = df[df['year'] == 2024]

# Set up colors for the pie chart
colors = ['blue', 'red', 'green', 'orange', 'purple', 'brown', 'pink', 'gray', 'cyan', 'magenta']

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(df_2024.values[0][1:], labels=df_2024.columns[1:], colors=colors, autopct='%1.1f%%')
plt.title('Population Distribution by Country in 2024')
plt.show()
```

### Stack bar chart:

```python
# here we make stackbar for check population increase in top 10 countries
# Filter data for the years 2024 and 2050
df_2024 = df[df['year'] == 2024]
df_2050 = df[df['year'] == 2050]

# Set up colors for the bars
color_2024 = 'blue'
color_2050 = 'orange'

# Plotting the stacked bar chart
plt.figure(figsize=(12, 8))

# Bar for 2024 population
plt.bar(df_2024.columns[1:], df_2024.values[0][1:], color=color_2024, label='2024')

# Bar for 2050 population stacked on top of 2024 population
plt.bar(df_2050.columns[1:], df_2050.values[0][1:], color=color_2050, bottom=df_2024.values[0][1:], label='2050')

# Annotations for 2024 population
for i, val in enumerate(df_2024.values[0][1:]):
    plt.text(i, val + 0.05 * val, str(val), ha='center', va='bottom')

# Annotations for 2050 population
for i, val in enumerate(df_2050.values[0][1:]):
    plt.text(i, df_2024.values[0][i+1] + val + 0.05 * val, str(val), ha='center', va='bottom')

plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Population Comparison for 2024 and 2050')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

```

### Heat map of 2024 and 2050:

```python
# here we going to check heatmap of 2024 and 2050

# Filter data for the years 2024 and 2050
df_2024 = df[df['year'] == 2024]
df_2050 = df[df['year'] == 2050]

# Concatenate data for both years
df_combined = pd.concat([df_2024, df_2050])

# Transpose the DataFrame for plotting
df_combined_transposed = df_combined.set_index('year').T

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df_combined_transposed, cmap='coolwarm', annot=True, fmt='d')
plt.title('Population Heatmap for 2024 and 2050')
plt.xlabel('Year')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

### Line chart:

```python
# here we are goint to make line chart for countries
# Plotting
plt.figure(figsize=(10, 6))
for column in df.columns[1:]:
    plt.plot(df['year'], df[column], marker='o', label=column.capitalize())

plt.title('Population Growth Over Years')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.xticks(df['year'], rotation=45)
plt.tight_layout()
plt.show()

```
